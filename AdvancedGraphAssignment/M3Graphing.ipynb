```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dfC = pd.read_csv('ConstructionTimeSeriesDatav2.csv')
```

```python
dfC.head()
```

```python
fig,ax = plt.subplots()

#line plots
ax.plot(dfC['Month-Year'], dfC['Private Construction'], label = "Private Construction")
ax.plot(dfC['Month-Year'], dfC['Public Construction'], label = "Public Construction")
ax.legend(loc=1, fontsize='14')

#titles and labels
fig.suptitle("Private and Public Spending on Construction Over Time",fontsize='18',fontname='Times New Roman')
plt.xlabel('Date (Month-Year)', fontsize='14', fontname='Times New Roman')
plt.ylabel('Total Construction Spending ($)', fontsize='14', fontname='Times New Roman')

#axes
plt.xticks(["Jan-02","Jan-03","Jan-04","Jan-05","Jan-06","Jan-07","Jan-08","Jan-09","Jan-10","Jan-11","Jan-12","Jan-13","Jan-14"])
ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction='in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction='in')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

#other fiugre settings
ax.axvspan(0,12,ymin=0,ymax = 10000,facecolor='k',alpha=0.1)
fig.set_size_inches(15,6)

plt.show()
```

### Intended Focus of the "Private and Public Spending on Construction Over TIme" Graph
***

I used shading on the graph to highlight that both private and public construction follow a 12-month, or annual, cycle. This is also known as seasonal variation. This may prompt the audience to estimate or better understand future private and public contruction trends.
